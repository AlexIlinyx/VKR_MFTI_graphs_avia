#### Предварительная обработку текста с пмощью spacy, сохранение в новые файлы txt

```python
import spacy

nlp = spacy.load("ru_core_news_sm")
# nlp = spacy.load("en_core_web_sm")

def preprocess_text(input_file_path, output_file_path):
    # Чтение текста из файла
    with open(input_file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Приведение текста к нижнему регистру
    text = text.lower()
    
    # Применение языковой модели spacy
    doc = nlp(text)
    
    # Удаление стоп-слов и знаков препинания, а также лемматизация
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    
    # Соединение токенов обратно в строку
    processed_text = ' '.join(tokens)
    
    # Запись обработанного текста в новый файл
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(processed_text)

# Пути к файлам
input_file_path = r"C:\Users\User\Documents\MFTI_VKR\__parsing\1\51176.txt"
output_file_path = r"C:\Users\User\Documents\MFTI_VKR\__parsing\1\51176-1.txt"

# Обработка текста и запись результата
preprocess_text(input_file_path, output_file_path)

# Вывод сообщения об успешном завершении
print("Текст был успешно обработан и сохранен в новом файле:", output_file_path)
```

#### Анализ отдельного файла Yandex gpt

```python
# Чтение текстового файла
file_path = r"C:\Users\User\Documents\MFTI_VKR\__parsing\1\51176-1.txt"
with open(file_path, 'r') as file:
    text = file.read()

# Определение размера блока для чтения файла
chunk_size = 1024 * 10  # 10 KB

# Отправляем файл на обработку частями
with open(file_path, 'r') as file:
    while True:
        chunk = file.read(chunk_size)
        if not chunk:
            break

        # Текст для обработки моделью Yandex GPT
        prompt = {
            "modelUri": f"gpt://{catalog_id}/yandexgpt-lite/latest",
            "completionOptions": {
                "stream": False,
                "temperature": 0.6,
                "maxTokens": "6000"
            },
            "messages": [
                {
                    "role": "user",
                    "text": chunk
                },
            ]
        }

        # Преобразование prompt в JSON
        prompt_json = json.dumps(prompt)
        print(prompt_json)
        # Отправляем POST-запрос к Yandex GPT API
        response = requests.post(url, headers=headers, data=prompt_json)

        # Проверка статуса ответа
        if response.status_code == 200:
            # Парсинг JSON-ответа
            result = response.json()

            # Проверка наличия ключа "result" в ответе
            if "result" in result:
                # Получение списка альтернативных сообщений
                alternatives = result["result"]["alternatives"]

                # Извлечение сущностей из каждой альтернативы
                for alternative in alternatives:
                    # Проверка наличия ключа "message" в альтернативе
                    if "message" in alternative:
                        # Извлечение сущностей из сообщения
                        entities = alternative["message"].get("entities", [])
                        # Вывод извлеченных сущностей
                        for entity in entities:
                            print(f"Entity: {entity['text']}, Type: {entity['type']}")
            else:
                print("No 'result' key found in the response.")
        else:
            print(f"Error: {response.status_code} - {response.reason}")

```
